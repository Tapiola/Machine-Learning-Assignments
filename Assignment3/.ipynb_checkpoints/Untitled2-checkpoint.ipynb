```python
from sklearn import datasets
from sklearn import svm
#from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
import keras
#from keras.utils.np_utils import to_categorical
np.random.seed(7)
train = pd.read_csv("fashion-mnist_train_small.csv").as_matrix()
Y = train[:, 0:1].copy()
#Y = np_utils.to_categorical(Y);
#.ravel()
X = train[:, 1:785].copy()
test = pd.read_csv("fashion-mnist_test_small.csv").as_matrix()
Y_test = test[:, 0:1].copy()
X_test = test[:, 1:785].copy()
# X = X.reshape(12000,28,28,1)
# X_test = X_test.reshape(3000,28,28,1)
#X, X_test, Y, Y_test = train_test_split(X, Y, test_size=0.1, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X)
X_train_std = sc.transform(X)
X_test_std = sc.transform(X_test)
```

```python
X_train_std = X_train_std.reshape(12000,28,28,1)
X_test_std = X_test_std.reshape(3000,28,28,1)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Conv2D, MaxPooling2D, Activation, Flatten, Dropout
model = Sequential()
model.add(Conv2D(32, (3, 3), input_shape=(28,28,1)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2), strides=(2, 2)))
model.add(Conv2D(32, (3, 3), input_shape=(28,28,1)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2), strides=(2, 2)))
model.add(Conv2D(64, (3, 3), padding="same"))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2), strides=(2, 2)))
model.add(Flatten())
model.add(Dense(512))
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(256))
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(10, activation='softmax'))
```

```python
from keras import losses
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(X_train_std, Y, validation_split=0.2, epochs=16, batch_size=50)
```

```python
scores = model.evaluate(X_test_std, Y_test)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

```python
model_json = model.to_json()
with open("model.json", "w") as json_file:
    json_file.write(model_json)
# serialize weights to HDF5
model.save_weights("model.h5")
```

```python
from matplotlib.colors import ListedColormap
import matplotlib.pyplot as plt
plt.figure(figsize=[8,6])
plt.plot(history.history['acc'],'r',linewidth=2.5)
plt.plot(history.history['val_acc'],'b',linewidth=2.5)
plt.legend(['Training Accuracy', 'Validation Accuracy'],fontsize=18)
plt.xlabel('Epochs ',fontsize=16)
plt.ylabel('Accuracy',fontsize=16)
plt.title('Accuracy Curves',fontsize=16)

plt.figure(figsize=[8,6])
plt.plot(history.history['loss'],'r',linewidth=2.5)
plt.plot(history.history['val_loss'],'b',linewidth=2.5)
plt.legend(['Training loss', 'Validation Loss'],fontsize=18)
plt.xlabel('Epochs ',fontsize=16)
plt.ylabel('Loss',fontsize=16)
plt.title('Loss Curves',fontsize=16)

plt.show()
```
